```python
vehicle_df_1['date-time'] = pd.to_datetime(vehicle_df_1['date-time'])
vehicle_df_2['date-time'] = pd.to_datetime(vehicle_df_2['date-time'])
vehicle_df_3['date-time'] = pd.to_datetime(vehicle_df_3['date-time'])
vehicle_df_4['date-time'] = pd.to_datetime(vehicle_df_4['date-time'])
vehicle_df_5['date-time'] = pd.to_datetime(vehicle_df_5['date-time'])

vehicle_df_1_60['date-time'] = pd.to_datetime(vehicle_df_1_60['date-time'])
vehicle_df_2_60['date-time'] = pd.to_datetime(vehicle_df_2_60['date-time'])
vehicle_df_3_60['date-time'] = pd.to_datetime(vehicle_df_3_60['date-time'])
vehicle_df_4_60['date-time'] = pd.to_datetime(vehicle_df_4_60['date-time'])
vehicle_df_5_60['date-time'] = pd.to_datetime(vehicle_df_5_60['date-time'])

vehicle_df_1_100['date-time'] = pd.to_datetime(vehicle_df_1_100['date-time'])
vehicle_df_2_100['date-time'] = pd.to_datetime(vehicle_df_2_100['date-time'])
vehicle_df_3_100['date-time'] = pd.to_datetime(vehicle_df_3_100['date-time'])
vehicle_df_4_100['date-time'] = pd.to_datetime(vehicle_df_4_100['date-time'])
vehicle_df_5_100['date-time'] = pd.to_datetime(vehicle_df_5_100['date-time'])
```

```python
# Aggregate from miliseconds to seconds.
vehicle_df_1 = vehicle_df_1.set_index('date-time').resample('1S').mean()
vehicle_df_2 = vehicle_df_2.set_index('date-time').resample('1S').mean()
vehicle_df_3 = vehicle_df_3.set_index('date-time').resample('1S').mean()
vehicle_df_4 = vehicle_df_4.set_index('date-time').resample('1S').mean()
vehicle_df_5 = vehicle_df_5.set_index('date-time').resample('1S').mean()

vehicle_df_1_60 = vehicle_df_1_60.set_index('date-time').resample('1S').mean()
vehicle_df_2_60 = vehicle_df_2_60.set_index('date-time').resample('1S').mean()
vehicle_df_3_60 = vehicle_df_3_60.set_index('date-time').resample('1S').mean()
vehicle_df_4_60 = vehicle_df_4_60.set_index('date-time').resample('1S').mean()
vehicle_df_5_60 = vehicle_df_5_60.set_index('date-time').resample('1S').mean()

vehicle_df_1_100 = vehicle_df_1_100.set_index('date-time').resample('1S').mean()
vehicle_df_2_100 = vehicle_df_2_100.set_index('date-time').resample('1S').mean()
vehicle_df_3_100 = vehicle_df_3_100.set_index('date-time').resample('1S').mean()
vehicle_df_4_100 = vehicle_df_4_100.set_index('date-time').resample('1S').mean()
vehicle_df_5_100 = vehicle_df_5_100.set_index('date-time').resample('1S').mean()
```

```python
# Create a seconds column for number of seconds spent each tick.
vehicle_df_1['Seconds'] = np.arange(len(vehicle_df_1))+1
vehicle_df_2['Seconds'] = np.arange(len(vehicle_df_2))+1
vehicle_df_3['Seconds'] = np.arange(len(vehicle_df_3))+1
vehicle_df_4['Seconds'] = np.arange(len(vehicle_df_4))+1
vehicle_df_5['Seconds'] = np.arange(len(vehicle_df_5))+1

vehicle_df_1_60['Seconds'] = np.arange(len(vehicle_df_1_60))+1
vehicle_df_2_60['Seconds'] = np.arange(len(vehicle_df_2_60))+1
vehicle_df_3_60['Seconds'] = np.arange(len(vehicle_df_3_60))+1
vehicle_df_4_60['Seconds'] = np.arange(len(vehicle_df_4_60))+1
vehicle_df_5_60['Seconds'] = np.arange(len(vehicle_df_5_60))+1

vehicle_df_1_100['Seconds'] = np.arange(len(vehicle_df_1_100))+1
vehicle_df_2_100['Seconds'] = np.arange(len(vehicle_df_2_100))+1
vehicle_df_3_100['Seconds'] = np.arange(len(vehicle_df_3_100))+1
vehicle_df_4_100['Seconds'] = np.arange(len(vehicle_df_4_100))+1
vehicle_df_5_100['Seconds'] = np.arange(len(vehicle_df_5_100))+1
```

```python
# Change in distance
vehicle_df_1['total_distance_travelled'] = vehicle_df_1['distanceOfTravel(meters)'].diff()
vehicle_df_2['total_distance_travelled'] = vehicle_df_2['distanceOfTravel(meters)'].diff()
vehicle_df_3['total_distance_travelled'] = vehicle_df_3['distanceOfTravel(meters)'].diff()
vehicle_df_4['total_distance_travelled'] = vehicle_df_4['distanceOfTravel(meters)'].diff()
vehicle_df_5['total_distance_travelled'] = vehicle_df_5['distanceOfTravel(meters)'].diff()

vehicle_df_1_60['total_distance_travelled'] = vehicle_df_1_60['distanceOfTravel(meters)'].diff()
vehicle_df_2_60['total_distance_travelled'] = vehicle_df_2_60['distanceOfTravel(meters)'].diff()
vehicle_df_3_60['total_distance_travelled'] = vehicle_df_3_60['distanceOfTravel(meters)'].diff()
vehicle_df_4_60['total_distance_travelled'] = vehicle_df_4_60['distanceOfTravel(meters)'].diff()
vehicle_df_5_60['total_distance_travelled'] = vehicle_df_5_60['distanceOfTravel(meters)'].diff()

vehicle_df_1_100['total_distance_travelled'] = vehicle_df_1_100['distanceOfTravel(meters)'].diff()
vehicle_df_2_100['total_distance_travelled'] = vehicle_df_2_100['distanceOfTravel(meters)'].diff()
vehicle_df_3_100['total_distance_travelled'] = vehicle_df_3_100['distanceOfTravel(meters)'].diff()
vehicle_df_4_100['total_distance_travelled'] = vehicle_df_4_100['distanceOfTravel(meters)'].diff()
vehicle_df_5_100['total_distance_travelled'] = vehicle_df_5_100['distanceOfTravel(meters)'].diff()
```

```python
def F(drag_coefficient, velocity, reference_area):
    return 0.5*1.225*drag_coefficient*reference_area*(velocity**2)
```

```python
#E = F Ã— d
def E(F, distance_of_travel_meters):
    return F * distance_of_travel_meters
```

```python
def E_input (k, E):
    return E / k
```

```python
vehicle_df_1["F"] = F(0.1, vehicle_df_1["VelocityMagnitude(BETA)"], 2.16)
vehicle_df_2["F"] = F(0.1, vehicle_df_2["VelocityMagnitude(BETA)"], 2.16)
vehicle_df_3["F"] = F(0.1, vehicle_df_3["VelocityMagnitude(BETA)"], 2.16)
vehicle_df_4["F"] = F(0.1, vehicle_df_4["VelocityMagnitude(BETA)"], 2.16)
vehicle_df_5["F"] = F(0.1, vehicle_df_5["VelocityMagnitude(BETA)"], 2.16)

vehicle_df_1_60["F"] = F(0.1, vehicle_df_1_60["VelocityMagnitude(BETA)"], 2.16)
vehicle_df_2_60["F"] = F(0.1, vehicle_df_2_60["VelocityMagnitude(BETA)"], 2.16)
vehicle_df_3_60["F"] = F(0.1, vehicle_df_3_60["VelocityMagnitude(BETA)"], 2.16)
vehicle_df_4_60["F"] = F(0.1, vehicle_df_4_60["VelocityMagnitude(BETA)"], 2.16)
vehicle_df_5_60["F"] = F(0.1, vehicle_df_5_60["VelocityMagnitude(BETA)"], 2.16)

vehicle_df_1_100["F"] = F(0.1, vehicle_df_1_100["VelocityMagnitude(BETA)"], 2.16)
vehicle_df_2_100["F"] = F(0.1, vehicle_df_2_100["VelocityMagnitude(BETA)"], 2.16)
vehicle_df_3_100["F"] = F(0.1, vehicle_df_3_100["VelocityMagnitude(BETA)"], 2.16)
vehicle_df_4_100["F"] = F(0.1, vehicle_df_4_100["VelocityMagnitude(BETA)"], 2.16)
vehicle_df_5_100["F"] = F(0.1, vehicle_df_5_100["VelocityMagnitude(BETA)"], 2.16)
```

```python
vehicle_df_1["E"] = E(vehicle_df_1['F'], vehicle_df_1['total_distance_travelled'])
vehicle_df_2["E"] = E(vehicle_df_2['F'], vehicle_df_2['total_distance_travelled'])
vehicle_df_3["E"] = E(vehicle_df_3['F'], vehicle_df_3['total_distance_travelled'])
vehicle_df_4["E"] = E(vehicle_df_4['F'], vehicle_df_4['total_distance_travelled'])
vehicle_df_5["E"] = E(vehicle_df_5['F'], vehicle_df_5['total_distance_travelled'])

vehicle_df_1_60["E"] = E(vehicle_df_1_60['F'], vehicle_df_1_60['total_distance_travelled'])
vehicle_df_2_60["E"] = E(vehicle_df_2_60['F'], vehicle_df_2_60['total_distance_travelled'])
vehicle_df_3_60["E"] = E(vehicle_df_3_60['F'], vehicle_df_3_60['total_distance_travelled'])
vehicle_df_4_60["E"] = E(vehicle_df_4_60['F'], vehicle_df_4_60['total_distance_travelled'])
vehicle_df_5_60["E"] = E(vehicle_df_5_60['F'], vehicle_df_5_60['total_distance_travelled'])

vehicle_df_1_100["E"] = E(vehicle_df_1_100['F'], vehicle_df_1_100['total_distance_travelled'])
vehicle_df_2_100["E"] = E(vehicle_df_2_100['F'], vehicle_df_2_100['total_distance_travelled'])
vehicle_df_3_100["E"] = E(vehicle_df_3_100['F'], vehicle_df_3_100['total_distance_travelled'])
vehicle_df_4_100["E"] = E(vehicle_df_4_100['F'], vehicle_df_4_100['total_distance_travelled'])
vehicle_df_5_100["E"] = E(vehicle_df_5_100['F'], vehicle_df_5_100['total_distance_travelled'])
```

```python
vehicle_df_1["energy_input(kWh)"] = E_input(0.88, vehicle_df_1['E'])/3.6e+6
vehicle_df_2["energy_input(kWh)"] = E_input(0.88, vehicle_df_2['E'])/3.6e+6
vehicle_df_3["energy_input(kWh)"] = E_input(0.88, vehicle_df_3['E'])/3.6e+6
vehicle_df_4["energy_input(kWh)"] = E_input(0.88, vehicle_df_4['E'])/3.6e+6
vehicle_df_5["energy_input(kWh)"] = E_input(0.88, vehicle_df_5['E'])/3.6e+6

vehicle_df_1_60["energy_input(kWh)"] = E_input(0.88, vehicle_df_1_60['E'])/3.6e+6
vehicle_df_2_60["energy_input(kWh)"] = E_input(0.88, vehicle_df_2_60['E'])/3.6e+6
vehicle_df_3_60["energy_input(kWh)"] = E_input(0.88, vehicle_df_3_60['E'])/3.6e+6
vehicle_df_4_60["energy_input(kWh)"] = E_input(0.88, vehicle_df_4_60['E'])/3.6e+6
vehicle_df_5_60["energy_input(kWh)"] = E_input(0.88, vehicle_df_5_60['E'])/3.6e+6

vehicle_df_1_100["energy_input(kWh)"] = E_input(0.88, vehicle_df_1_100['E'])/3.6e+6
vehicle_df_2_100["energy_input(kWh)"] = E_input(0.88, vehicle_df_2_100['E'])/3.6e+6
vehicle_df_3_100["energy_input(kWh)"] = E_input(0.88, vehicle_df_3_100['E'])/3.6e+6
vehicle_df_4_100["energy_input(kWh)"] = E_input(0.88, vehicle_df_4_100['E'])/3.6e+6
vehicle_df_5_100["energy_input(kWh)"] = E_input(0.88, vehicle_df_5_100['E'])/3.6e+6
```
